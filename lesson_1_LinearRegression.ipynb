```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update({'figure.figsize':(7,3),'figure.dpi':120}) #cài đặt kích thước cho toàn bộ các hình trong file này giống nhau

#from google.colab import files
#uploaded = files.upload()

data = pd.read_csv('Data_Lesson1_ab_Startups.csv')

data.head()
data.info()
data.describe()

#Bài 1a) Xây dựng mô hình hồi quy tuyến tính với 1 biến (Linear Regression one variable)

x = data['RnD']
y = data['Profit']

#vẽ hình thể hiện hai biến x và y
plt.scatter(data.RnD, data.Profit, color = 'purple', marker = 'o')
plt.xlabel('Chi phí RnD (1000$)',fontsize=12)
plt.ylabel('Chi phí lợi nhuận (1000$)',fontsize=12)
plt.grid()

x=data[['RnD']]
y=data['Profit'].values

#xây dựng mô hình
from sklearn.linear_model import LinearRegression

#1 gọi mô hình từ thư viện gán vào mô hình một biến (model)
model_linear = LinearRegression()

#2 Training model (model.fit(x,y))
model_linear.fit(x,y)

a = model_linear.coef_
b = model_linear.intercept_

#In ra thông tin của mô hình y = ax + b với a là độ dốc coef_ và b là intercept
print('a = ',a)
print('b = ',b)

#Phương trình hồi quy tuyến tình cần tìm là y = 0.8x + 54

model_linear.predict([[542]])

#3 Dự báo
# model_linear.predict([[542]]) => Dự đoán với giá trị x bất kì thì giá trị y bằng bao nhiêu. ([[100]]) là biểu thị cho kiểu dữ liệu
y_predict = model_linear.predict(data[['RnD']]) # dự đoán lại với chuỗi y
y_predict

c=model_linear.predict([[1000]])
print(c)

# vẽ đường dự đoán cùng với hình biểu diễn hai biến x và y
plt.scatter(data.RnD, data.Profit, color = 'orange', marker = 'o', label = 'Dữ liệu thực tế')
plt.plot(data.RnD, y_predict, color = 'blue', label = 'y dự báo')
plt.xlabel('Chi phí RnD (1000$)',fontsize=12)
plt.ylabel('Chi phí lợi nhuận (1000$)',fontsize=12)
plt.legend()
plt.grid()

#4 Tính toán độ chính xác (r^2)
from sklearn.metrics import r2_score
ac = r2_score(data.Profit, y_predict)
print('The accuracy of this model is R-square = ',"{:.2f}".format(ac))
```

```python
#Bài tập 1: Thực hành xây dựng mô hình tính lợi nhuận (y) dựa trên chi phí Quản lý (x) (Adm)
x1 = data['Adm']
y1 = data['Profit']

plt.scatter(data.Adm,data.Profit,color='red',marker='*')
plt.xlabel('Chi phí quản lí (1000$)',fontsize=12)
plt.ylabel('Chi phí lợi nhuận (1000$)',fontsize=12)
plt.grid()

x1=data[['Adm']]
y1=data['Profit'].values

model_linear_1 = LinearRegression()

model_linear_1.fit(x1,y1)

a1 = model_linear_1.coef_
b1 = model_linear_1.intercept_

print('a1 = ',a1)
print('b1 = ', b1)

y_predict_1 = model_linear_1.predict(data[['Adm']])
y_predict_1

plt.scatter(data.Adm, data.Profit, color = 'red', marker='*',label='Dữ liệu thực tế')
plt.plot(data.Adm, y_predict_1, color = 'green',label = 'y1 dự báo')
plt.xlabel('Chi phí quản lí (1000$)',fontsize=12)
plt.ylabel('Chi phí lợi nhuận (1000$)',fontsize=12)
plt.legend()
plt.grid()

ac1 = r2_score(data.Profit, y_predict_1)
print('The accuracy of this model is R-square = ',"{:.4f}".format(ac1))
```
