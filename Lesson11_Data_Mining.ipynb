```python
# import thư viện
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams.update({'figure.figsize':(7,3), 'figure.dpi':120})
```

```python
link='https://drive.google.com/file/d/1Ei9FXfzS6MV9qF-74pof3ZM1tu8oMqhT/view?usp=sharing'
# Nguồn data https://raw.githubusercontent.com/HuynhHoa99/AprioriAlgorithm/master/Day1.csv
#Đọc data
path = 'https://drive.google.com/uc?export=download&id='+link.split('/')[-2]
df = pd.read_csv(path,encoding= 'unicode_escape')
```

```python
df
```

```python
#Cài đặt
!pip install apyori
from apyori import apriori
```

```python
assciation_rules = apriori(df.values, min_support=0.50,min_lift=1.1,min_confidence=0.7)
```

```python
# chúng ta có thể thay đổi cài đặt về  min_support, min_confidenc và min_lift để khám phá các quy luật khác nhau
```

```python
results = list(assciation_rules)
```

```python
len(results)
```

```python
for i in range (len(results)):
    print(results[i])
```

```python
# đưa kết quả về 1 bảng để tiện đọc
df1 = 0
df1 = pd.DataFrame(columns=['Itemset','Support','Items_base','Items_add','Confidence','Lift'])
#Lặp qua các kết quả của Results và đặt vào bảng
for i in range (len(results)):
    for record in results:
        items = record[0]
        support = record[1]
        ordered_statistics = record[2]
        for stat in ordered_statistics:
            items_base = stat[0]
            items_add = stat[1]
            confidence = stat[2]
            lift = stat[3]
            # đưa kết quả vào bảng
            df1 = df1.append({'Itemset': items, 'Support': support, 'Items_base' : items_base,
                              'Items_add': items_add, 'Confidence':confidence, 'Lift':lift},
                              ignore_index=True)
```

HƯỚNG DẪN ĐỌC KẾT QUẢ:
- items_base: là đối tượng đọc trước (A)
- items_add: là đối tượng đọc sau (B)

Thực hành với các mức độ min_support và min_confidence khác nhau và nhận xét

Ý nghĩa của các chỉ số
1. support: sự kết hợp này có phổ biến hay không?
2. confidence:  đo mức độ ảnh hưởng của hành động A (items_base) tới hành động B (items_add)
Lưu ý:hành động hặc nhóm hành động
3. Lift: Tương quan giữa hành động A và B
